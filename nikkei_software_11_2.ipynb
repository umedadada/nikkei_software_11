```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.figure_factory as ff
from numpy.ma.core import sqrt
from scipy.stats import t
from scipy import stats
```

```python
df=pd.read_csv('ramen_score.csv')
```

```python
df.head()
```

データの可視化

```python
hist_data = [df.query("group=='osaka'")['score'], 
             df.query("group=='tokyo'")['score']]
group_labels = ['osaka', 'tokyo']
fig = ff.create_distplot(hist_data, group_labels, bin_size=.1)
fig.show()
```

```python
# 標本平均の算出
tokyo_score_avg = df[df['group']=='tokyo']['score'].mean()
osaka_score_avg = df[df['group']=='osaka']['score'].mean()

print('東京の平均スコア：{}，大阪の平均スコア：{}'.format(tokyo_score_avg, osaka_score_avg))
```

定義上の計算

```python
# それぞれのスコアを抽出
tokyo_score = df[df['group']=='tokyo']['score']
osaka_score = df[df['group']=='osaka']['score']
```

```python
# 必要な統計量を算出する関数を作成
def calc_stats(score):
  # 母平均の推定
  x_bar=score.mean()

  # 平方和の計算
  s=0
  for x in score:
    s += (x  - x_bar )**2

  # サンプル数の計算
  n=len(score)
  return x_bar,s,n
```

```python
# それぞれの群の統計量を計算
x_bar1,s1,n1 = calc_stats(tokyo_score)
x_bar2,s2,n2 = calc_stats(osaka_score)

print(x_bar1,s1,n1,x_bar2,s2,n2)
```

```python
# 検定統計量の計算
## 分散の計算
var = (s1 + s2) / (n1 + n2 - 2)
## T値の計算
t0=(x_bar1 - x_bar2) / np.sqrt(var * ( 1/n1 + 1/n2))
## 得られたT値をもとに下側確率(p値)の計算
p_value = t.cdf(x=t0,df=n1 + n2 - 2)

print(var,t0,p_value)
```

ライブラリを使った場合

```python
stats.ttest_ind(tokyo_score, osaka_score, alternative='less')
```
